```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import streamlit as st
from datetime import datetime
import re

# Load the data
print("Loading metadata.csv...")
df = pd.read_csv('metadata.csv')

# Examine the first few rows and data structure
print("First 5 rows of the dataset:")
print(df.head())

print("\nData structure (info):")
print(df.info())

# Basic data exploration
print(f"\nDataFrame dimensions: {df.shape[0]} rows, {df.shape[1]} columns}")

print("\nData types of each column:")
print(df.dtypes)

print("\nMissing values in important columns:")
important_cols = ['title', 'abstract', 'publish_time', 'journal', 'authors']
print(df[important_cols].isnull().sum())

print("\nBasic statistics for numerical columns:")
print(df.describe())
# Handle missing data
print("Columns with many missing values:")
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent.sort_values(ascending=False).head(10))

# Create a cleaned version of the dataset
# For analysis, we'll keep rows that have at least title and publish_time
df_cleaned = df.dropna(subset=['title', 'publish_time']).copy()

# Fill missing abstract with empty string
df_cleaned['abstract'] = df_cleaned['abstract'].fillna('')

# Fill missing journal with 'Unknown'
df_cleaned['journal'] = df_cleaned['journal'].fillna('Unknown')

# Prepare data for analysis
# Convert date columns to datetime format
df_cleaned['publish_time'] = pd.to_datetime(df_cleaned['publish_time'], errors='coerce')

# Extract year from publication date
df_cleaned['year'] = df_cleaned['publish_time'].dt.year

# Create new columns
df_cleaned['abstract_word_count'] = df_cleaned['abstract'].apply(lambda x: len(str(x).split()))
df_cleaned['title_word_count'] = df_cleaned['title'].apply(lambda x: len(str(x).split()))

print(f"Cleaned dataset dimensions: {df_cleaned.shape[0]} rows, {df_cleaned.shape[1]} columns}")
# Perform basic analysis
# Count papers by publication year
year_counts = df_cleaned['year'].value_counts().sort_index()

# Identify top journals publishing COVID-19 research
top_journals = df_cleaned['journal'].value_counts().head(10)

# Find most frequent words in titles
def get_word_frequencies(text_series, n=20):
    all_text = ' '.join(text_series.dropna().astype(str).str.lower())
    words = re.findall(r'\b[a-zA-Z]+\b', all_text)
    word_freq = pd.Series(words).value_counts()
    # Remove common stop words
    stop_words = ['the', 'and', 'of', 'in', 'to', 'a', 'for', 'with', 'on', 'by', 'as', 'an', 'at', 'from', 'that', 'is', 'are', 'this', 'be', 'we', 'using']
    word_freq = word_freq[~word_freq.index.isin(stop_words)]
    return word_freq.head(n)

title_word_freq = get_word_frequencies(df_cleaned['title'])

# Create visualizations
plt.figure(figsize=(15, 10))

# Plot number of publications over time
plt.subplot(2, 2, 1)
year_counts.plot(kind='bar')
plt.title('Number of Publications by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Create a bar chart of top publishing journals
plt.subplot(2, 2, 2)
top_journals.plot(kind='bar')
plt.title('Top 10 Journals by Publication Count')
plt.xlabel('Journal')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Generate a word cloud of paper titles
plt.subplot(2, 2, 3)
all_titles = ' '.join(df_cleaned['title'].dropna().astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_titles)
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Paper Titles')
plt.axis('off')

# Plot distribution of paper counts by source
plt.subplot(2, 2, 4)
source_counts = df_cleaned['source_x'].value_counts().head(10)
source_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 10 Sources by Publication Count')
plt.ylabel('')

plt.tight_layout()
plt.savefig('visualizations.png')
plt.show()

print("Top words in titles:")
print(title_word_freq)
```
